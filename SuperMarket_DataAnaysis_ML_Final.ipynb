# Treating with outliers

37. To Find out the ouliers in the profit column and create a new data frame without the outliers ?

1. Using Z Score method

```python
#visualize the outlier for profit
import seaborn as sns
sns.boxplot(q1['Profit'])
```

```python
#detect the outliers using the z score 
outliers=[]
dd1=q1['Profit']
def det_outliers(dd1):
  threshold = 2
  mean = np.mean(dd1)
  std = np.std(dd1)
  
  for i in dd1:
    z_score = (i-mean)/std
    if z_score> threshold:
      outliers.append(i)  
  return outliers
```

```python
out = det_outliers(dd1)
```

```python
len(out)
```

2. Using Quartile range method

```python
import numpy as np
q1,q3=np.percentile(out,[25,75])
```

```python
print(q1,q3)
```

```python
iqr = q3-q1
iqr
```

```python
lowerfen = q1-(1.5*iqr) # finding out the lower fence and the higher fence
higherfen = q3+(1.5*iqr)
print(lowerfen,higherfen)
```

```python
import sklearn
import pandas as pd
# IQR
Q1 = np.percentile(data['Profit'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(data['Profit'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", data.shape)
 
# Upper bound
upper = np.where(data['Profit'] >= (Q3+1.5*IQR))
lower = np.where(data['Profit'] <= (Q1-1.5*IQR))
 
#Removing the Outliers
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)
 
print("New Shape: ", data.shape)
```
